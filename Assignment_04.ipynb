### Que1.-
Predict if a person would buy life insurance based on his age using logistic regression.
About Dataset - Age: Age of a person
bought_insurance : whether person buys Insurance or not (0 - No, 1 - Yes)
predict it for Age : 58,30,20,49,53

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

### Dataset : insurance_data.csv

```python
df = pd.read_csv("insurance_data.csv")
X = df[['age']]
y = df['bought_insurance']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
```

### Train model

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

### Predictions

```python
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]
```

### Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

### Accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)
```

### Precision

```python
precision = precision_score(y_test, y_pred)
print("Precision Score:", precision)
```

### Recall

```python
recall = recall_score(y_test, y_pred)
print("Recall Score:", recall)
```

### F1 Score

```python
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)
```

### roc_auc Score

```python
roc_auc = roc_auc_score(y_test, y_prob)
print("ROC AUC Score:", roc_auc)
```

## Prediction

```python


def Prediction_CHD():
    male=int(input("Enter male(1)/female(0):"))
    age=int(input("Age:"))
    edu=int(input("Enter Education(Some high school (1), high school/GED (2), some college/vocational school (3), college (4)):"))
    currentSmoker=int(input("whether or not the patient is a current smoker (0=No, 1=Yes):"))
    cigsPerDay=int(input("the number of cigarettes that the person smoked on average in one day:"))
    BPMeds=int(input("whether or not the patient was on blood pressure medication (0):"))
    prevalentStroke=int(input("whether or not the patient had previously had a stroke (0=No, 1=Yes):"))
    prevalentHyp=int(input("whether or not the patient was hypertensive (0=No, 1=Yes):"))
    diabetes=float(input("whether or not the patient had diabetes (0=No, 1=Yes):"))
    totChol=float(input("total cholesterol level:"))
    sysBP=float(input("systolic blood pressure "))
    diaBP=float(input("diastolic blood pressure"))
    BMI=float(input("Body Mass Index"))
    heartRate=float(input("heart rate"))
    glucose=float(input("glucose level"))

    data=[[male,age,edu,currentSmoker,cigsPerDay,BPMeds,prevalentStroke,prevalentHyp,diabetes,totChol,sysBP,diaBP,BMI,heartRate,glucose]]

    pred=model.predict(data)

    print(f"10-year risk of Coronary Heart Disease (CHD)(Yes(1)/NO(0)):{pred}")
```
