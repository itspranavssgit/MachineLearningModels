```python
import numpy as np
```

```python
a1 = np.array([10, 20, 30, 40, 50]) 
a2 = np.array([10, 20, 30, 40, 50], dtype=np.int32) 
a3 = np.array([10, 20, 30, 40, 50], dtype=np.int16) 
a4 = np.array([10, 20, 30, 40, 50], dtype=np.int8)
```

```python
a1 = np.array([10, 20, 30, 40, 50]) 
print(a1.dtype) 
# int64
```

```python
a1 = np.array([10, 20, 30, 40, 50]) 
print(a1.size)
```

```python
a1 = np.array([10, 20, 30, 40, 50]) 
print(a1.itemsize) # 8 
a2 = np.array([10, 20, 30, 40, 50], dtype=np.int32) 
print(a2.itemsize) # 4 
a3 = np.array([10, 20, 30, 40, 50], dtype=np.int16) 
print(a3.itemsize) # 2 
a4 = np.array([10, 20, 30, 40, 50], dtype=np.int8) 
print(a4.itemsize)
```

```python
a1 = np.array([10, 20, 30, 40, 50]) 
print(a1.nbytes)
```

```python
a1 = np.array([10, 20, 30, 40, 50]) 
print(a1.ndim) # 1 
a2 = np.array([[10, 20], [30, 40]]) 
print(a2.ndim)
```
